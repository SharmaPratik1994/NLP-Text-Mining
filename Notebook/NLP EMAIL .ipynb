## Stop Words Removal

```python
def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    filtered_tokens = [word for word in text if word.lower() not in stop_words]
    return ' '.join(filtered_tokens)
```

```python
aggregated_data['cleaned_tokens'] = aggregated_data['cleaned_tokens'].apply(remove_stopwords)
```

```python
aggregated_data['cleaned_tokens']
```

## Stemming

```python
from nltk.stem import PorterStemmer

stemmer = PorterStemmer()
```

```python
def stem_tokens(tokens):
    return [stemmer.stem(token) for token in tokens]
```

```python
aggregated_data['stemmed_tokens'] = aggregated_data['cleaned_tokens'].apply(stem_tokens)
```

```python
aggregated_data['stemmed_tokens']
```

```python
# Convert tokenized documents to list of strings
aggregated_data['tokenized_docs_as_strings'] = [' '.join(tokens) for tokens in aggregated_data['stemmed_tokens']]
```

```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')
# Initialize lemmatizer
lemmatizer = WordNetLemmatizer()

# Lemmatize the tokens in the 'cleaned_tokens' column
aggregated_data['lemmatized_tokens'] = aggregated_data['cleaned_tokens'].apply(
    lambda tokens: [lemmatizer.lemmatize(token, pos=wordnet.VERB) for token in tokens]
)
```
